```python
#######Problem 3#########
import csv
import pandas as pd
import numpy as np

##Reads into a pandas DataFrame
df = pd.read_csv("https://raw.githubusercontent.com/CompTools/Class_Files/master/data/CO-OPS__8729108__wl.csv",
                header = 0)
##Add column headers
df.columns = ["Date_Time","Water_Level","Sigma","O","F","R","L","Quality"]

##Create a new variable that is the difference between one water level and the on right above it 
df['Diff'] = df["Water_Level"] - df["Water_Level"].shift(1)

##returns the max value for the new diff variable and returns information from that row including time and date
df.loc[df["Diff"].idxmax()]

```
